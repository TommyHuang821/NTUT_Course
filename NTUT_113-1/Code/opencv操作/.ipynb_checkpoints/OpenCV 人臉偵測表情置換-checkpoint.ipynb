## 3. 人臉置換

```python



def get_face_landmarks(image, face_detector, shape_predictor):
    """
    GET 68 face landmarkers
    :param image: image
    :param face_detector: dlib.get_frontal_face_detector
    :param shape_predictor: dlib.shape_predictor
    :return: np.array([[],[]]), 68 face landmarkers
    """
    dets = face_detector(image, 1)
    num_faces = len(dets)
    if num_faces == 0:
        print("Sorry, there were no faces found.")
        return None
    shape = shape_predictor(image, dets[0])
    face_landmarks = np.array([[p.x, p.y] for p in shape.parts()])
    return face_landmarks


def get_face_mask(image_size, face_landmarks):
    """
    Get face mask
    :param image_size: image size
    :param face_landmarks: 68 face landmarkers
    :return: image_mask, 
    """
    # mask = np.zeros(image_size, dtype=np.uint8)
    # points = np.concatenate([face_landmarks[0:16], face_landmarks[26:17:-1]])
    # cv2.fillPoly(img=mask, pts=[points], color=255)

    mask = np.zeros(image_size, dtype=np.uint8)
    points = cv2.convexHull(face_landmarks)  # 凸包
    cv2.fillConvexPoly(mask, points, color=255)
    return mask


def get_affine_image(image1, image2, face_landmarks1, face_landmarks2):
    """
    Get AffineTransform image1
    :param image1: for AffineTransform
    :param image2: need iamge2 imagesize for image 1 project to image2 image size
    :param face_landmarks1: image 1 facelandmarker
    :param face_landmarks2: image 2 facelandmarker
    :return: AffineTransform image1
    """
    three_points_index = [18, 8, 25]
    M = cv2.getAffineTransform(face_landmarks1[three_points_index].astype(np.float32),
                               face_landmarks2[three_points_index].astype(np.float32))
    dsize = (image2.shape[1], image2.shape[0])
    affine_image = cv2.warpAffine(image1, M, dsize)
    return affine_image.astype(np.uint8)


def get_mask_center_point(image_mask):
    """
    get center point for a mask image
    :param image_mask: 
    :return: center point
    """
    image_mask_index = np.argwhere(image_mask > 0)
    miny, minx = np.min(image_mask_index, axis=0)
    maxy, maxx = np.max(image_mask_index, axis=0)
    center_point = ((maxx + minx) // 2, (maxy + miny) // 2)
    return center_point


def get_mask_union(mask1, mask2):
    """
    get two covers' union 
    :param mask1: mask_image, cover1
    :param mask2: mask_image, cover2
    :return: union
    """
    mask = np.min([mask1, mask2], axis=0)  # 掩盖部分并集
    mask = ((cv2.blur(mask, (7, 7)) == 255) * 255).astype(np.uint8)  # 缩小掩模大小
    mask = cv2.blur(mask, (5, 5)).astype(np.uint8)  # 模糊掩模
    return mask


def skin_color_adjustment(im1, im2, mask=None):
    """
    skin color adjustment
    :param im1: image 1
    :param im2: image 2
    :param mask: face mask. 如果存在，使用人脸部分平均值来求skin color變換係數；其他，使用Gaussian smooth来求skin color變換係數
    :return: adjust image1 by image2 color
    """
    if mask is None:
        im1_ksize = 55
        im2_ksize = 55
        im1_factor = cv2.GaussianBlur(im1, (im1_ksize, im1_ksize), 0).astype(np.float64)
        im2_factor = cv2.GaussianBlur(im2, (im2_ksize, im2_ksize), 0).astype(np.float64)
    else:
        im1_face_image = cv2.bitwise_and(im1, im1, mask=mask)
        im2_face_image = cv2.bitwise_and(im2, im2, mask=mask)
        im1_factor = np.mean(im1_face_image, axis=(0, 1))
        im2_factor = np.mean(im2_face_image, axis=(0, 1))
    im1 = np.clip((im1.astype(np.float64) * im2_factor / np.clip(im1_factor, 1e-6, None)), 0, 255).astype(np.uint8)
    return im1




```

```python
# -*- coding: utf-8 -*-

import os
import cv2
import dlib
import numpy as np

image_face_path = os.path.join( 'JayChou.png')  # reference

predictor_path = os.path.join( 'shape_predictor_68_face_landmarks.dat')  

detector = dlib.get_frontal_face_detector()  
predictor = dlib.shape_predictor(predictor_path)  

def main():
    im1 = cv2.imread(image_face_path)  # face_image
    im1 = cv2.resize(im1, (600, im1.shape[0] * 600 // im1.shape[1]))
    landmarks1 = get_face_landmarks(im1, detector, predictor)  # 68_face_landmarks
    if landmarks1 is None:
        print('{}:no face detected'.format(image_face_path))
        exit(1)
    im1_size = (im1.shape[0], im1.shape[1]) # image1 image size
    im1_mask = get_face_mask(im1_size, landmarks1)  # face mask

    cam = cv2.VideoCapture(0)
    while True:
        ret_val, im2 = cam.read()  # camera_image
        landmarks2 = get_face_landmarks(im2, detector, predictor)  # 68_face_landmarks
        if landmarks2 is not None:
            im2_size = (im2.shape[0], im2.shape[1]) # image2 image size
            im2_mask = get_face_mask(im2_size, landmarks2)  # face mask

            affine_im1 = get_affine_image(im1, im2, landmarks1, landmarks2)  # im1 affine transformation
            affine_im1_mask = get_affine_image(im1_mask, im2, landmarks1, landmarks2)  # im1 mask affine transformation

            union_mask = get_mask_union(im2_mask, affine_im1_mask)  # 
            affine_im1 = skin_color_adjustment(affine_im1, im2, mask=union_mask)  # skin color adjustment
            
            point = get_mask_center_point(affine_im1_mask)  # 
            seamless_im = cv2.seamlessClone(affine_im1, im2, mask=union_mask, p=point, flags=cv2.NORMAL_CLONE)  # face fusion
            
            
            combined_image = cv2.hconcat([im2, seamless_im])
            cv2.imshow('seamless_im', combined_image)
        else:
            cv2.imshow('seamless_im', im2)
        if cv2.waitKey(1) == 27:  # 按Esc退出
            break
    cv2.destroyAllWindows()


if __name__ == '__main__':
    main()
```
