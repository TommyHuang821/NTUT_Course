# 高斯雜訊

假設X是高斯雜訊，
$$
X\sim N(0,\sigma^2)
$$

最重要的特性是**平均數是0**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

time        = np.arange(0, 10, 0.1);
signal   = np.sin(time)

signal_noise =  signal + 0.1*np.random.randn(len(time))

plt.plot(time, signal)
plt.title('sine wave')
plt.show()

plt.plot(time, signal_noise)
plt.title('sine wave with noise (N(0, 0.1))')
plt.show()

def sinwithnoise(signal, n):
    # 假設有N筆帶有雜訊的sine wave合成
    for i in range(n):
        if i==0:
            signal_noises =  signal + 0.1*np.random.randn(len(time))
        else:
            signal_noises+= signal + 0.1*np.random.randn(len(time))
    signal_noises/=n
    return signal_noises


for i, n in enumerate([5,10,50]):
    signal_noises = sinwithnoise(signal, n)
    mse = np.sum(np.abs(signal_noises-signal))/len(time)
    plt.plot(time, signal_noises)
    plt.title('{} data: error={}'.format(n, mse))
    plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

time    = np.arange(0, 10, 0.1);
event   = np.sin(time)*0.1

signal = np.random.randn(len(time))*0.001

signal[40:60] += event[40:60]


signal_noise =  signal + 0.1*np.random.randn(len(time))

plt.plot(time, signal)
plt.title('signal')
plt.show()

plt.plot(time, signal_noise)
plt.title('signal with noise (N(0, 0.1))')
plt.show()

def sinwithnoise(signal, n):
    # 假設有N筆帶有雜訊的sine wave合成
    for i in range(n):
        if i==0:
            signal_noises =  signal + 0.1*np.random.randn(len(time))
        else:
            signal_noises+= signal + 0.1*np.random.randn(len(time))
    signal_noises/=n
    return signal_noises


for i, n in enumerate([10,50,500]):
    signal_noises = sinwithnoise(signal, n)
    mse = np.sum(np.abs(signal_noises-signal))/len(time)
    plt.plot(time, signal_noises)
    plt.title('{} data: error={}'.format(n, mse))
    plt.show()

```
