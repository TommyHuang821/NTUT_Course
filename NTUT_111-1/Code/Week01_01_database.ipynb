## 從sklearn讀取資料庫

### IRIS資料庫

<img src="Image/IRIS.png" width="60%">

```python
from sklearn.datasets import load_iris
data_iris = load_iris()
print(data_iris.keys())
print('Number of IRIS data: {}.\nDimension:{}.'.format(data_iris.data.shape[0],data_iris.data.shape[1]))
```

```python
# 印出第0,60,140筆data
print(data_iris.data[[0, 60, 140],:])
```

```python
# 印出第0,60,140筆label
print(data_iris.target[[0, 60, 140]])
```

```python
print(data_iris.target_names)
```

### 波士頓房價預測 資料庫

### 如果不想要看整個資料集的class，可以只讀取 data(X) 和 target(Y)

範例: 以波士頓房價為例子

只需要在load data內加入「return_X_y=True」即可，如下

```python
X, Y = load_boston(return_X_y=True)
print(X[[0],:])
print(Y[[0]])
```

```python
X, Y = load_iris(return_X_y=True)
print(X[[0],:])
print(Y[[0]])
```

## 不只是結構資料，非結構資料sklearn也有內建的
### 數字資料庫
此資料庫為1797筆8x8大小的影像所組成，每張圖為手寫的數字，如下圖




```python
from sklearn.datasets import load_digits
data_digits = load_digits()
print('Number of Digits data: {}.\nDimension:{}.'.format(data_digits.data.shape[0],data_digits.data.shape[1]))
```

#### 手寫數字視覺化

雖然data_digits.data是用64維的array儲存圖片，但其實8x8圖片也被儲存在data_digits.images內

```python
data_digits.images.shape
```

```python
import matplotlib.pyplot as plt
_, axes = plt.subplots(nrows=1, ncols=10, figsize=(20, 10))
for ax, image, label in zip(axes, data_digits.images, data_digits.target):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('label: %i' % label)
```

### 還有一些圖內建於sklearn

```python
from sklearn.datasets import load_sample_images
dataset = load_sample_images()  
dataset
_, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 10))
for ax, image in zip(axes, dataset.images):
    ax.set_axis_off()
    ax.imshow(image)
```
