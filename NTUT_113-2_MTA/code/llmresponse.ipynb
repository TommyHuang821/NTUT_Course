## åˆ©ç”¨OPENAI: whisperä¾†é€²è¡ŒèªéŸ³è½‰æ–‡å­—

```python
import torch
import whisper
import os

verbose = True  #æ˜¯å¦å³æ™‚é¡¯ç¤ºèªéŸ³è¾¨è­˜çµæœ
whisper_model_type = 'base' # tiny,base,small,medium,large,large-v2,large-v3
lang = 'Chinese' #["Chinese", "English", "Japanese", "Korean"]
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
model = whisper.load_model(whisper_model_type, device=DEVICE)

audioFile = 'å¹¹è©±èªéŒ„.mp3'
file_exist = os.path.isfile(audioFile)
if file_exist:
    print('processing: {}'.format(audioFile))
    result = model.transcribe(audioFile, fp16=False, verbose=verbose, language=lang)
else:
    print('file not found')
```

```python
import pyttsx3
engine = pyttsx3.init()
# ç²å–æ‰€æœ‰å¯ç”¨èªéŸ³
voices = engine.getProperty('voices')
for index, voice in enumerate(voices):
    print(f"Voice {index}: {voice.name} - {voice.languages} - {voice.id}")
```

```python
# èªéŸ³è¼¸å‡º
engine = pyttsx3.init()
engine.setProperty('voice', voices[0].id)  # å¯æ”¹ç´¢å¼•ä¾†æ›´æ›èªéŸ³
engine.setProperty("rate", 300)  # è¨­å®šèªé€Ÿ
engine.setProperty("volume", 0.6)  # è¨­å®šéŸ³é‡
text_read = 'Hello å¤§å®¶å¥½ï¼Œæ­¡è¿ä¾†åˆ°åœ‹ç«‹å°åŒ—ç§‘æŠ€å¤§å­¸ 113å­¸å¹´åº¦ç¬¬2å­¸æœŸ å¤šåª’é«”æŠ€è¡“èˆ‡æ‡‰ç”¨'
# ä½¿ç”¨ TTS è®€å‡ºå›æ‡‰
engine.say(text_read)
engine.runAndWait()
```
